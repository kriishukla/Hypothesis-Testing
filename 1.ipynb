```python
import csv

def load_csv_to_2d_list(filename):
    data = []
    with open(filename, mode='r', newline='') as file:
        reader = csv.reader(file)
        for row in reader:
            data.append(row)
    return data

```

```python
def find_discrete_attributes(data):
    discrete_attributes = []
    num_rows = len(data)
    
    if num_rows == 0:
        return discrete_attributes
    
    num_cols = len(data[0])
    
    for col in range(num_cols):
        unique_values = set()
        for row in data[1:]:  
            unique_values.add(row[col])

        if len(unique_values) < 10: 
            discrete_attributes.append(col)
    
    return discrete_attributes
```

```python
def calculate_mean(data):
    n = len(data)
    mean_vector = sum(data) / n
    return mean_vector

```

```python
def calculate_variance(data, mean_vector):
    n = len(data)
    variance = sum((xi - mean_vector) ** 2 for xi in data) / n
    return variance
```

```python
def normalize_data(df, numeric_columns):
    normalized_df = df.copy()
    for col in numeric_columns:
        mean = normalized_df[col].mean()
        variance = normalized_df[col].var(ddof=0)  
        std_dev = variance ** 0.5  
        normalized_df[col] = (normalized_df[col] - mean) / std_dev
    return normalized_df
```

```python
def chi_square_test(contingency_table):
    observed = contingency_table.values
    total = sum(sum(row) for row in observed)
    
    row_sums = [sum(row) for row in observed]
    col_sums = [sum(observed[i][j] for i in range(len(observed))) for j in range(len(observed[0]))]

    chi_square_stat = 0
    for i in range(len(observed)):
        for j in range(len(observed[0])):
            expected = (row_sums[i] * col_sums[j]) / total
            chi_square_stat += (observed[i][j] - expected) ** 2 / expected

    return chi_square_stat
```

```python


input_filename = '/home/krishna/Downloads/DSC/Assignment-1 - Auto MPG.csv'
df_encoded = pd.read_csv(input_filename)

contingency_table = pd.crosstab(df_encoded['model year'], df_encoded['cylinders']) 

chi_square_stat = chi_square_test(contingency_table)

df = (len(contingency_table.index) - 1) * (len(contingency_table.columns) - 1)


alpha = 0.05

critical_value = 5.991  


```

```python
print(f"Chi-Square Statistic: {chi_square_stat:.4f}")
print(f"Degrees of Freedom: {df}")
print(f"Critical Value: {critical_value:.4f}")

if chi_square_stat > critical_value:
    print("Reject the null hypothesis: Model year has an effect on the number of cylinders.")
else:
    print("Fail to reject the null hypothesis: Model year does not have an effect on the number of cylinders.")
```
